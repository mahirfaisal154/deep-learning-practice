```python
for values in df['TotalCharges'].unique():
    print(values)
df['TotalCharges']=pd.to_numeric(df['TotalCharges'])
df['TotalCharges']=df['TotalCharges'].replace(np.nan,0)
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer
nz=Normalizer()
sc=StandardScaler()
columns_to_scale=['tenure','MonthlyCharges','TotalCharges']
df[columns_to_scale]=nz.fit_transform(df[columns_to_scale])
df.sample(10)
df.dtypes
```

```python
df.dtypes
input_dataset = pd.concat([df.iloc[:, 0:3], df.iloc[:, 5:]], axis=1)
output_dataset=df.iloc[:,4]
output_dataset.head
```

```python
output_dataset.replace({"Yes":1,"No":0},inplace=True)
output_dataset.head
```

```python

from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_dataset,output_dataset,test_size=.2,random_state=5)
```

```python
from tensorflow import keras
from keras import layers
models=keras.Sequential([
    keras.layers.Dense(44,input_shape=(44,),activation='sigmoid'),
    keras.layers.Dense(32,activation='sigmoid'),
    keras.layers.Dense(16,activation='sigmoid'),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(4,activation='relu'),
    keras.layers.Dense(1,activation='relu')
   ])
models.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'],)
models.fit(x_train,y_train,epochs=250)
                        
```

```python
yp=models.predict(x_test)
from sklearn.metrics import accuracy_score
for i in range(len(yp)):
    if yp[i]>.5:
        yp[i]=1
    else:
        yp[i]=0
result=accuracy_score(y_test,yp)
result

            
    
        
```
